```python
_characters = "ءابتثجحخدذرزسشصضطظعغفقلمنهويِپچژکگیآۀأؤإئًَُّ!(),-.:;?  ̠،…؛؟‌٪#ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz_–@+/\\\u200c"

def clean_text_verbose(text):
    allowed_chars = set(_characters)
    removed = set(c for c in text if c not in allowed_chars)
    cleaned = ''.join(c for c in text if c in allowed_chars)
    if removed:
        print(f"🚫 Removed characters: {' '.join(sorted(removed))}")
    return cleaned

# Example usage
raw_text = "این یک متن آزمایشی با کاراکترهای نادرست ♥🔥😊 و مجاز است."
cleaned_text = clean_text_verbose(raw_text)
print("🧼 Cleaned text:", cleaned_text)
```

```python
import pysrt
from pydub import AudioSegment
import os
import subprocess


def stretch_audio_to_fit(input_wav, output_wav, target_duration_ms):
    """Adjusts audio speed using FFmpeg to match a target duration."""
    original = AudioSegment.from_wav(input_wav)
    original_duration = len(original)
    speed_factor = original_duration / target_duration_ms

    # Clamp speed factor to range for quality
    speed_factor = max(0.5, min(speed_factor, 2.0))

    if abs(speed_factor - 1.0) < 0.01:
        os.rename(input_wav, output_wav)
        return

    print(f"⏩ Stretching audio: {input_wav} -> {output_wav} | Speed factor: {speed_factor:.2f}")

    try:
        subprocess.run(
            [
                "ffmpeg", "-y", "-loglevel", "error",
                "-i", input_wav,
                "-filter:a", f"atempo={speed_factor}",
                output_wav
            ],
            check=True
        )
    except subprocess.CalledProcessError:
        print(f"❌ FFmpeg failed while processing: {input_wav}")


def get_total_duration_from_subs(subs):
    last_end = subs[-1].end
    return (
        last_end.hours * 3600 +
        last_end.minutes * 60 +
        last_end.seconds
    ) * 1000 + last_end.milliseconds


def load_or_initialize_audio(path, duration_ms):
    if os.path.exists(path):
        print("🔄 Loaded partial audio from previous run.")
        return AudioSegment.from_wav(path)
    else:
        print("🎬 Starting new combined audio.")
        return AudioSegment.silent(duration=duration_ms)


def read_checkpoint(path):
    if os.path.exists(path):
        with open(path, 'r') as f:
            return int(f.read().strip())
    return -1


def write_checkpoint(path, index):
    with open(path, 'w') as f:
        f.write(str(index))


# === Config ===
srt_path = 'translated.srt'
checkpoint_file = 'progress.txt'
partial_audio_path = 'combined_audio_partial.wav'
ref_wav_path = '/content/Persian-Tacotron2-on-ManaTTS/sample.wav'
script_path = '/content/Persian-MultiSpeaker-Tacotron2/inference.py'
cwd_path = '/content/Persian-MultiSpeaker-Tacotron2'

# === Load subtitles and setup ===
subs = pysrt.open(srt_path, encoding='utf-8')
total_duration = get_total_duration_from_subs(subs)
combined = load_or_initialize_audio(partial_audio_path, total_duration)
last_index = read_checkpoint(checkpoint_file)

print(f"🔁 Resuming from subtitle index {last_index + 1}")

# === Process subtitles ===
for i, sub in enumerate(subs):
    if i <= last_index:
        continue

    try:
        # Time window in ms
        start_time = (sub.start.hours * 3600 + sub.start.minutes * 60 + sub.start.seconds) * 1000 + sub.start.milliseconds
        end_time = (sub.end.hours * 3600 + sub.end.minutes * 60 + sub.end.seconds) * 1000 + sub.end.milliseconds
        time_window = end_time - start_time

        raw_text = sub.text.strip().replace('\n', ' ')
        input_text = clean_text_verbose(raw_text)
        chunk_path = f"chunk_{i + 1}.wav"

        print(f"🎤 Generating audio for subtitle {i}: {input_text}")

        command = [
            "python3", script_path,
            "--vocoder", "HiFiGAN",
            "--text", input_text,
            "--ref_wav_path", ref_wav_path,
            "--test_name", f"chunk_{i + 1}"
        ]

        subprocess.run(command, cwd=cwd_path, check=True)

        output_path = f"/content/Persian-MultiSpeaker-Tacotron2/results/chunk_{i + 1}.wav"

        chunk = AudioSegment.from_wav(output_path)
        if len(chunk) > time_window:
            print(f"⚠️ Warning: {chunk_path} is too long by {(len(chunk) - time_window) / 1000:.2f} sec")
            stretched_path = f"chunk_stretched_{i + 1}.wav"
            stretch_audio_to_fit(output_path, stretched_path, time_window)
            chunk = AudioSegment.from_wav(stretched_path)

        combined = combined.overlay(chunk, position=start_time)

        # Save progress
        write_checkpoint(checkpoint_file, i)
        combined.export(partial_audio_path, format="wav")

    except Exception as e:
        print(f"❌ Error at subtitle {i}: {e}")
        print(f"💥 Last completed index: {i - 1}")
        print(f"➡️ To resume from here, run this in a new cell:")
        print(f"with open('{checkpoint_file}', 'w') as f:\n    f.write('{i - 1}')")
        break

# === Final export ===
combined.export("combined_audio.wav", format="wav")
print("✅ Final combined audio saved as 'combined_audio.wav'")
```
